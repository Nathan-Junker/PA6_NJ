```python
import math
class calc():
    def __init__(self):
        pass
        #elf.input1=input1
        #elf.input2=input2
    
    def addition(num1, num2):
        print(f"Addition= {num1 +num2}\n")
    
    def subtraction(num1, num2):
        print(f"Subtraction= {num1 - num2}\n")
        
    def multiplication(num1, num2):
        print(f"Multiplication= {num1 * num2}\n")
        
    def division(num1, num2):
        print(f"Division = {num1 / num2}\n")
        
    def exponentiation (num1, num2):
        print(f"Exponentiation = {num1**num2}")
        
    def sin(num1):
        print(f"Sin= {math.sin(num1)}")
    def cos(num1):
        print(f"Cos= {math.cos(num1)}")
            
    def tan(num1):
        print(f"Tan= {math.tan(num1)}")
        
    def logarithm(num1):
        if num1 > 0:
            print(f"Log= {math.log10(num1)}")
        else:
            print("Domain Error")
    
    
    def natural_logarithm(num1):
        if num1 >0:
            print(f"ln= {math.log(num1, math.e)}")
        else:
            print("Domain Error")
            
    def exp(num1):
        print(f"Exp= {(math.e)**num1}")
        
    def arcsin(num1):
        if num1 <=1 and num1>=0:
            print(f"arcsin= {math.asin(num1)}")
        else:
            print("Domain Error")
            
    def arccos(num1):
        if num1 <=1 and num1>=0:
            print(f"arccos= {math.acos(num1)}")
        else:
            print("Domain Error")
    def arctan(num1):
        print(f"arctan= {math.atan(num1)}")
    
    def fraction_to_decimal(num1,num2):
        whole_number= int(round(num1//num2))
        remainder = (num1/num2)-whole_number
        print(f'Fraction to decimal= {whole_number+remainder}')
        
        
while True:
    print("Choose from the following menu: ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exponentiation")
    print("6. Sin")
    print("7. Cos")
    print("8. Tan")
    print("9. Log")
    print("10. ln")
    print("11. Exp")
    print("12. arcsin")
    print("13. arccos")  
    print("14. arctan")
    print("15. Fraction to Decimal")
    print("16. Exit")
    
    choice=int(input("Enter your choice (1-16): "))
    if choice<1 or choice>16:
        print("wrong choice")
    
    elif choice== 1:
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        calc.addition(num1,num2)
    elif choice== 2:
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        calc.subtraction(num1,num2)
    elif choice==3:
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        calc.multiplication(num1,num2)
    elif choice==4:
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        if num2== 0:
            print("Infinity")
        else:
            calc.division(num1,num2)
            
    elif choice == 5: 
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        if num1 and num2 != 0:
            calc.exponentiation(num1,num2)
        else:
            print("Error")
    
    elif choice == 6:
        num1=float(input("Enter 1st number: "))
        calc.sin(num1)
    
    elif choice == 7:
        num1=float(input("Enter 1st number: "))
        calc.cos(num1)
        
    elif choice== 8:
        num1=float(input("Enter 1st number: "))
        calc.tan(num1)
        
    elif choice== 9:
        num1=float(input("Enter 1st number: "))
        calc.logarithm(num1)
        
    elif choice== 10:
        num1=float(input("Enter 1st number: "))
        calc.natural_logarithm(num1)
        
    elif choice==11:
        num1=float(input("Enter 1st number: "))
        calc.exp(num1)
    
    elif choice== 12:
        num1=float(input("Enter 1st number: "))
        calc.arcsin(num1)
        
    elif choice== 13:
        num1=float(input("Enter 1st number: "))
        calc.arccos(num1)
        
    elif choice== 14:
        num1=float(input("Enter 1st number: "))
        calc.arctan(num1)
        
    elif choice== 15:
        num1=float(input("Enter 1st number: "))
        num2=float(input("Enter 2nd number: "))
        calc.fraction_to_decimal(num1,num2)
        
    elif choice == 16:
        break
    
    print("\n")
    
```
